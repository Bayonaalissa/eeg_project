```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
features_train = h2o.upload_file("data/dwt_features_train.csv",header=1)
features_test = h2o.upload_file("data/dwt_features_test.csv",header=1)
```

```python

# Identify predictors and response
x = features_train.columns
y = "720"
x.remove(y)


# For binary classification, response should be a factor
features_train[y] = features_train[y].asfactor()
features_test[y] = features_test[y].asfactor()
# Run AutoML for 20 base models

aml = H2OAutoML(sort_metric='AUCPR', seed=1)
aml.train(x=x, y=y, training_frame=features_train)

# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

```python
best_model = aml.get_best_model(criterion='mean_per_class_error')
```

```python
best_performance = best_model.model_performance(features_test)
```

```python
best_performance.accuracy()
```

```python
best_performance.confusion_matrix()
```
